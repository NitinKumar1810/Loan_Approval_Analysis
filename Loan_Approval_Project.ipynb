# Handling Null Values

```python
# Filling null values in Gender column

df.Gender = df.Gender.fillna(df.Gender.mode()[0])
```

```python
# Filling null values in Dependents column

df.Dependents = df.Dependents.fillna(df.Dependents.mode()[0])
```

```python
# Filling null values in Self_Employed column

df.Self_Employed = df.Self_Employed.fillna(df.Self_Employed.mode()[0])
```

```python
# Filling null values in LoanAmount column

df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.median())
```

```python
# Filling null values in Loan_Amount_Term column

df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.median())
```

```python
# Filling null values in Credit_History column

df.Credit_History = df.Credit_History.fillna(df.Credit_History.mode()[0])
```

```python
# lets check if there is any null values left or not

df.isnull().sum()
```

## Checking Outliers

```python
sns.boxplot(x=df.ApplicantIncome)
```

# Observation:
* The majority of the outliers are indentified, and they all exceeds 20000.

```python
df[df.ApplicantIncome>20000]
```

# Observation:
* After analyzing the 'ApplicantIncome' column, an outlier was identified with a substantially high value of 72529. This exception is causing a deviation from the straight pattern in the chart, a decision was made to treat this specific entry as an outlier.
* To mitigate the influence of the outlier, we replace this outlier with the upper limit. This adjustment is intended to align the corrected value more closely with the typical range of applicant incomes, facilitating a more accurate representation of the data for subsequent analyses.

```python
# Find Q1 and Q3
Q1 = df.ApplicantIncome.quantile(.25)
Q3 = df.ApplicantIncome.quantile(.75)
```

```python
Q1,Q3
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
# Finding Upper limit and Lower limit
UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
# Replacing the outliers with UL
df.ApplicantIncome = np.where(df.ApplicantIncome==72529,UL,df.ApplicantIncome)
```
